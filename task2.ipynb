```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
print(data.head())
print(data.info())
```

```python
data.columns = data.columns.str.strip()
data.dropna(inplace=True)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data)
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Group by Region and calculate average unemployment rate
state_unemployment = (
    data.groupby('Region')['Estimated Unemployment Rate (%)']
    .mean()
    .sort_values()
)

# Plot bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
state_unemployment.plot(kind='barh')
plt.title("Average Unemployment Rate by Region")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Region")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Estimated Labour Participation Rate (%)',
    y='Estimated Unemployment Rate (%)',
    data=data
)
plt.title("Labour Participation vs Unemployment Rate")
plt.xlabel("Labour Participation Rate (%)")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
